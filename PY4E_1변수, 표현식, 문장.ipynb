# Variables 변수
하나의 값만 가지기에
x = 12.2
x = 100
을 하면 순차적으로 실행하여
print(x)의 값은 100이 된다.

## Name Rules
* Must start with a letter or underscore (숫자로 시작할 수 없습니다.)
* Must consist of letters, numbers, and underscores
* 변수이름은 아무렇게나 해도 되지만 사람의 가독성을 위해 코드를 작성하기를 권장

```python
x = 12.2
print(x)
```

## 할당문 (Assignment Statements)
x = 3.9 * x * ( 1 - x )
라는 식이 있을 때 x값에 0.6을 넣는다면 우측을 먼저 계산하여 나온결과를 좌측x값에 넣는다.

```python
x= 0.6
x = 3.9 * x * (1 -x )
print(x)
```
